```python
# 遍历输入目录中的所有 .txt 文件
input_dir = r"D:\HKU\Corpus\CHAT"
# print(os.listdir(input_dir))
for filename in os.listdir(input_dir):
    print(filename)
    if filename.endswith(".txt"):
        input_path = os.path.join(input_dir, filename)

        # 读取文件内容
        with open(input_path, "r", encoding="utf-8-sig") as f:
            text = f.read()

        # 分词
        tokens = word_tokenize(text)

        # 词性标注
        tagged_tokens = pos_tag(tokens)
        # print(tagged_tokens)

        # # 将标注结果写回同一个文件
        with open(input_path, "w", encoding="utf-8") as f:
            for word, tag in tagged_tokens:
                if word not in string.punctuation:
                    # print(word, tag)
                    f.write(f"{word}_{tag} ")
```

```python
MOD = 10**18

def count_unique_strings(n, k):
    if k == 0:
        return 1
    
    dp = [0] * (k + 1)
    dp[0] = 1  # 初始状态：0次修改
    
    for _ in range(n):
        new_dp = [0] * (k + 1)
        # 前缀和数组用于优化偶数次修改的计算
        prefix = [0] * (k + 1)
        prefix[0] = dp[0]
        for j in range(1, k + 1):
            prefix[j] = (prefix[j - 1] + dp[j]) % MOD
        
        for j in range(k + 1):
            # 情况1：当前字符不修改，贡献1
            res = dp[j]
            
            # 情况2：当前字符修改奇数次（至少1次），贡献25
            if j >= 1:
                res = (res + dp[j - 1] * 25) % MOD
            
            # 情况3：当前字符修改偶数次（至少2次），贡献26
            if j >= 2:
                res = (res + prefix[j - 2] * 26) % MOD  # 前缀和优化
            
            new_dp[j] = res % MOD
        
        dp = new_dp
    
    return dp[k] % MOD

# 示例
n = 2
k = 2
print(count_unique_strings(n, k)) 
print(2*25*(24+25))
```
